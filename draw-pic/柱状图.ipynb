```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.font_manager import FontProperties
from matplotlib.pyplot import MultipleLocator,FormatStrFormatter
import matplotlib as mpl
datas = pd.read_excel('newdata.xlsx')
datas = datas.values
Edge = []
My = []
Node = []

for data in datas:
    if(data[0]=="GFAN" and data[3]==1):
        Edge.append(data[9])
        print(data[2])
    if(data[1]=="D_ablation_MyModel_TUAF"and data[3]==1):
        My.append(data[9])
    if(data[1]=="D_ablation_MyModel_w/o_SVDD"and data[3]==1):
        Node.append(data[9])
print(len(Edge))
fig = plt.figure(figsize=(18,3))
mpl.rcParams['text.color'] = 'black'
mpl.rcParams['font.family'] = 'Cambria Math'
ax1=plt.gca()
ax1.patch.set_facecolor("white")
ax1.tick_params(axis='x',colors='black')
ax1.tick_params(axis='y',colors='black')
ax1.spines['top'].set_color('black')  
ax1.spines['right'].set_color('black')  
ax1.spines['bottom'].set_color('black')
ax1.spines['left'].set_color('black')
ax1.margins(0.01)  
rect = fig.patch
rect.set_facecolor('white')
label = ['Edge','My','Node']
label_font1 = {'family': 'Cambria Math',
                  'size': 15,
                  'color':'black',
                  }
label_font = {'family': 'Cambria Math',
                  'size': 20,
                  'color':'black',
                  }
tick_label = ["PTC_MR","PTC_FM","PTC_FR","AIDS","ARE","MMP","HSE","AhR",'PPAR',"p53","PC-3","MCF-7","MOLT-4","UACC257","SN12C","SF-295","SW-620","P388"]

    
x = np.arange(len(Edge))


plt.xticks(x,tick_label,rotation=-35)

width = 0.15
plt.bar(x-0.3, Edge,width,color="#f8ac8c",label='w/o Reconstruction',edgecolor='black') #9AC9DB
plt.bar(x-0.15, Node,width,color="#B0C4DE",label='w/o One-class Classifier',edgecolor='black')#F8AC8C
plt.bar(x, My,width,color="#2878B5",label='TUAF',edgecolor='black')#C00000
plt.rc('font',family='Times New Roman')
# plt.legend(bbox_to_anchor=(0.02, 1.05), loc=3, borderaxespad=0,ncol=4,columnspacing=28.5,facecolor='white',frameon=False) #绘制图例
plt.legend(loc='best',bbox_to_anchor=(1,0.94),borderaxespad=0,ncol=4,facecolor='white',frameon=False,fontsize=16)
# plt.xlabel("Datasets", label_font1) 
plt.ylabel("AUC", label_font)
plt.tick_params(labelsize=16)

ax=plt.gca()

# plt.savefig('data.png'bbox_inches = 'tight')
plt.show()

```
